```python
for i in ['alpha', 'delta','redshift', 'plate', 'MJD']:
    plt.figure(figsize=(13,7))
    sb.histplot(data=star_data, x=i, kde=True, hue="class")
    plt.title(i)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
star_data["class"] = le.fit_transform(star_data["class"])
star_data["class"] = star_data["class"].astype(int)
    
def log_plot(column):
    for i in range(3):
        sb.kdeplot(data=np.log(star_data[star_data["class"] == i][column]), label = le.inverse_transform([i]), fill = True)
    sb.kdeplot(data=np.log(star_data[column]),label = ["All"], fill = True)
    plt.legend()
    plt.show()
```

```python
for i in ['u', 'g', 'r', 'i', 'z']:
    log_plot(i)
    
```

# Outliers deletion

```python
def rem_outliers():
    s1 = star_data.shape

    for i in star_data.select_dtypes(include = 'number').columns:
        qt1 = star_data[i].quantile(0.25)
        qt3 = star_data[i].quantile(0.75)
        iqr =  qt3 - qt1
        lower = qt1-(1.5*iqr)
        upper = qt3+(1.5*iqr)
        min_in = star_data[star_data[i]<lower].index
        max_in = star_data[star_data[i]>upper].index
        star_data.drop(min_in, inplace = True)
        star_data.drop(max_in, inplace = True)
    
    s2 = star_data.shape
    outliers = s1[0] - s2[0]
    return outliers
```

```python
print("Number of outliers deleted are : ", rem_outliers())
```
